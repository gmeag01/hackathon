```python
for i in range(len(os.listdir('./final_data/DB_10S_FL/')[:304])):
    st = 200
    end = st + 600

    FL_csv = pd.read_csv(os.path.join('./final_data/DB_10S_FL/', FL_list[i]))
    FR_csv = pd.read_csv(os.path.join('./final_data/DB_10S_FR/', FR_list[i]))
    HL_csv = pd.read_csv(os.path.join('./final_data/DB_10S_HL/', HL_list[i]))
    HR_csv = pd.read_csv(os.path.join('./final_data/DB_10S_HR/', HR_list[i]))
    W_csv = pd.read_csv(os.path.join('./final_data/DB_10S_W/', W_list[i]))

    # -------------------------------- norm ----------------------------------

    FL_scale = scaler.fit_transform(FL_csv[["ax", "ay", "az", "gx", "gy", "gz"]])
    FL_norm = pd.DataFrame(FL_scale, columns=[["ax", "ay", "az", "gx", "gy", "gz"]])

    FR_scale = scaler.fit_transform(FL_csv[["ax", "ay", "az", "gx", "gy", "gz"]])
    FR_norm = pd.DataFrame(FL_scale, columns=[["ax", "ay", "az", "gx", "gy", "gz"]])

    HL_scale = scaler.fit_transform(FL_csv[["ax", "ay", "az", "gx", "gy", "gz"]])
    HL_norm = pd.DataFrame(FL_scale, columns=[["ax", "ay", "az", "gx", "gy", "gz"]])

    HR_scale = scaler.fit_transform(FL_csv[["ax", "ay", "az", "gx", "gy", "gz"]])
    HR_norm = pd.DataFrame(FL_scale, columns=[["ax", "ay", "az", "gx", "gy", "gz"]])

    W_scale = scaler.fit_transform(FL_csv[["ax", "ay", "az", "gx", "gy", "gz"]])
    W_norm = pd.DataFrame(FL_scale, columns=[["ax", "ay", "az", "gx", "gy", "gz"]])

    for cnt in range(len(FL_norm)):
        if end + 50 * cnt > len(FL_csv):
            break
        fl.append(FL_norm.iloc[st + 50 * cnt : end + 50 * cnt])
        fr.append(FR_norm.iloc[st + 50 * cnt : end + 50 * cnt])
        hl.append(HL_norm.iloc[st + 50 * cnt : end + 50 * cnt])
        hr.append(HR_norm.iloc[st + 50 * cnt : end + 50 * cnt])
        w.append(W_norm.iloc[st + 50 * cnt : end + 50 * cnt])

        if '_01' in FL_list[i]:
            y.append(0)
        else:
            y.append(1)

fl_train = np.array(fl)
fr_train = np.array(fr)
hl_train = np.array(hl)
hr_train = np.array(hr)
w_train = np.array(w)
y_train = np.array(y)
```

# 모델 생성

```python
from tensorflow.python.keras import Model, Input
from tensorflow.python.keras.layers import Conv1D, MaxPooling1D, Concatenate, Dense, Dropout
from tensorflow.python.keras import losses
from tensorflow.keras.layers import BatchNormalization
```

```python
def CNN_dim(input):
    network = Conv1D(16, 6, activation='relu')(input)
    network = BatchNormalization()(network) 
    network = MaxPooling1D(2, 2)(network) 

    network = Conv1D(32, 6, activation='relu')(network)
    network = BatchNormalization()(network) 
    network = MaxPooling1D(2, 2)(network) 

    network = Conv1D(64, 6, activation='relu')(network)
    network = BatchNormalization()(network) 
    network = MaxPooling1D(2, 2)(network)

    network = Conv1D(128, 6, activation='relu')(network)
    network = BatchNormalization()(network) 
    network = MaxPooling1D(2, 2)(network)

    return network
```

```python
inputFL = Input(shape=(600, 6), name='FL_input')
inputFR = Input(shape=(600, 6), name='FR_input')
inputHL = Input(shape=(600, 6), name='HL_input')
inputHR = Input(shape=(600, 6), name='HR_input')
inputW = Input(shape=(600, 6), name='W_input')

cnnFL = CNN_dim(inputFL)
cnnFR = CNN_dim(inputFR)
cnnHL = CNN_dim(inputHL)
cnnHR = CNN_dim(inputHR)
cnnW = CNN_dim(inputW)

concat = Concatenate()([cnnFL, cnnFR, cnnHL, cnnHR, cnnW])
# lstm = (LSTM(256))(concat)

h = Dense(128, activation='relu')(concat)
h = Dense(64, activation='relu')(h)
h = Dropout(0.5)(h)
output = Dense(1, activation='sigmoid')(h)
```

```python
model = Model(inputs=[inputFL, inputFR, inputHL, inputHR, inputW], outputs=output)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', 
              loss=losses.BinaryCrossentropy(), 
              metrics=['accuracy'])
model.fit(x = [fl_train, fr_train, hl_train, hr_train, w_train],
          y = y_train,
          validation_data=([fl_test, fr_test, hl_test, hr_test, w_test], y_test),
          epochs=1,
          batch_size=32,
          verbose=1)
```
